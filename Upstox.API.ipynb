```python
import pandas as pd
import time
data = {
    'instrument_name': ['NIFTY', 'NIFTY', 'NIFTY', 'NIFTY'],
    'strike_price': [19500, 19500, 20000, 20000],
    'side': ['PE', 'CE', 'PE', 'CE'],
    'bid_price': [0.65, None, 1.2, None],
    'ask_price': [None, 2302.25, None, 2400.5]
}

# Create DataFrame
options_data = pd.DataFrame(data)

# Function 1: Get option chain data
def get_option_chain_data(instrument_name: str, expiry_date: str, side: str) -> pd.DataFrame:
    """
    Retrieves highest bid price for puts (PE) or highest ask price for calls (CE) for each strike.

    Parameters:
    - instrument_name (str): Name of the instrument (e.g., NIFTY).
    - expiry_date (str): Expiration date of the options.
    - side (str): Type of option to retrieve ("PE" for put, "CE" for call).

    Returns:
    - DataFrame with columns: instrument_name, strike_price, side, bid/ask.
    """
    # Filter for the given instrument, expiry date, and side
    filtered_data = options_data[(options_data['instrument_name'] == instrument_name) & (options_data['side'] == side)]

    # Select the relevant price column based on side
    price_column = 'bid_price' if side == 'PE' else 'ask_price'

    # Group by strike price and get the max bid or ask price
    result = filtered_data.groupby(['instrument_name', 'strike_price', 'side'])[price_column].max().reset_index()
    result.rename(columns={price_column: 'bid/ask'}, inplace=True)

    return result

# Function 2: Calculate margin and premium
def calculate_margin_and_premium(data: pd.DataFrame) -> pd.DataFrame:
    """
    Calculates margin required and premium earned for each option.

    Parameters:
    - data (DataFrame): DataFrame from get_option_chain_data with option details.

    Returns:
    - DataFrame with additional columns: margin_required, premium_earned.
    """
    # Set a predefined margin and lot size (these would typically come from an API or fixed values for demo)
    lot_size = 1200  # Example lot size
    margin_per_contract = 1950  # Example margin per contract for demonstration purposes

    # Calculate margin and premium for each row
    data['margin_required'] = margin_per_contract
    data['premium_earned'] = data['bid/ask'] * lot_size

    return data
def main():
# Record start time
    start_time = time.time()

    # Execute functions
    option_chain = get_option_chain_data(instrument_name="NIFTY", expiry_date="2025-12-26", side="PE")
    result_with_margin_premium = calculate_margin_and_premium(option_chain)

    # Record end time
    end_time = time.time()

    # Calculate execution time
    execution_time = end_time - start_time

    # Return result and execution time
    return result_with_margin_premium, execution_time

# Call main function and get results
result, exec_time = main()
print(result)
print("Execution Time:", exec_time)
```
